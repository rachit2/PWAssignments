```python
# 5. List vs Tuple mutability
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_list[0] = 100
# my_tuple[0] = 100  # Will raise TypeError
print(my_list)
print(my_tuple)
```

```python
# 6. Function for mutable and immutable arguments
def test_mutable_immutable(a, b):
    a += 5
    b.append(10)
    print(a, b)

x = 10
y = [1, 2]
test_mutable_immutable(x, y)
print(x, y)
```

```python
# 13. Sum of even numbers between 1 and 50
sum_even = 0
for i in range(2, 51, 2):
    sum_even += i
print(sum_even)
```

```python
# 14. Reverse string using while loop
text = input("Enter a string: ")
i = len(text) - 1
rev = ''
while i >= 0:
    rev += text[i]
    i -= 1
print(rev)
```

```python
# 15. Factorial using while loop
n = int(input("Enter a number: "))
fact = 1
while n > 0:
    fact *= n
    n -= 1
print(fact)
```
